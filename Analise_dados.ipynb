```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Estatísticas básicas
df.describe()
```

```
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='renda_mensal', y='gasto_mensal')
plt.title('Renda vs Gasto Mensal')
plt.show()
```
